```python
import numpy as np 
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs, make_circles

%matplotlib inline
```

```python
seed = 42;
```

```python
# 1.1 Plot blobs using sklearn functions

A, y1 = make_blobs(n_samples = 250, centers = [(-1,-1)], cluster_std = 1/8, random_state = seed)
B, y2 = make_blobs(n_samples = 250, centers = [(1,1)], cluster_std = 1/8, random_state = seed+1)

plt.scatter(A[:,0],A[:,1], marker = '.', c = 'b', alpha = 0.3)
plt.scatter(B[:,0],B[:,1], marker = '.', c = 'r', alpha = 0.3)
```

```python
# 1.2 Plot blobs using numpy functions
np.random.seed(seed)
A_new = np.random.randn(250,2)*1/8 + [-1,-1]
np.random.seed(seed+1)
B_new = np.random.randn(250,2)*1/8 + [1,1]
plt.scatter(A_new[:,0],A_new[:,1], marker = '.', c = 'b', alpha = 0.3)
plt.scatter(B_new[:,0],B_new[:,1], marker = '.', c = 'r', alpha = 0.3)
```

```python
# 1.3 Plot the blobs using a larger std

A, y1 = make_blobs(n_samples = 250, centers = [(-1,-1)], cluster_std = 4, random_state = seed)
B, y2 = make_blobs(n_samples = 250, centers = [(1,1)], cluster_std = 4, random_state = seed+1)

plt.scatter(A[:,0],A[:,1], marker = '.', c = 'b', alpha = 0.7)
plt.scatter(B[:,0],B[:,1], marker = '.', c = 'r', alpha = 0.7)
```

```python
# 1.4 Non linearly separable

x,ind = make_circles(n_samples = (500,500),noise = 0.2, random_state = seed+1)
B = x[ind>0]
A,y = make_blobs(n_samples = 500, centers = [(0,0)], cluster_std = 1/8, random_state = seed)
plt.scatter(A[:,0],A[:,1], marker = '.', c = 'b', alpha = 0.2)
plt.scatter(B[:,0],B[:,1], marker = '.', c = 'r', alpha = 0.5)
```
